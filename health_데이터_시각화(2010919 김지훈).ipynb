

#####**인사이트 1 정리**
(총콜레스테롤과 HDL 콜레스테롤간의 관계)


* 총콜레스테롤(total cholesterol, TC)은  일반적으로 저밀도지질단백질(LDL), 고밀도지질단백질(HDL), 중성지방의 수치를 합산한 콜레스테롤 수치이다.

* 트라이글리세라이드: 지방의 한 형태로써, 중성 지방이 이 종류에 속한다.

*   총콜레스테롤과 다음 3가지 요소(LDL, HDL, 트라이글리세라이드)의 관계를 살펴본 결과, 총콜레스테롤과 LDL이 **양의 관계**를 보임을 확인할 수 있었다. 


1.   산점도와 선 그래프로 확인해본 결과, 총콜레스테롤과 HDL, 트라이글라세라이드는 유의미한 상관관계를 보이지 않는다.
2.   LDL콜레스테롤 수치가 높으면, 혈관 속 혈전을 발색시켜 혈관질환 위험이 커진다. 
3.   총콜레스테롤이 높은 사람일 경우, LDL 또한 높기에, 혈관 질환 위험이 높아진다. 
4.   총콜레스테롤이 높은 범위인 250에서 400사이를 기준으로 산점도를 확인하였을 때, 9, 10, 12, 13 그룹 나이대가 많이 확인된다. 즉, 40~60세 사이에서 총콜레스테롤과 LDL 콜레스테롤의 양의 관계가 더욱 잘 확인된다. 

###**인사이트 2: 흡연상태와 혈색소 간의 관계**

```python
sns.boxplot(data = health, x = '흡연상태', y = '혈색소')
#현재 흡연을 하고 있는 사람이 비흡연자(No), NotNow 보다 혈색소 수치가 높음
```

```python
sns.set(rc={'figure.figsize':(6, 5)}) 

sns.lineplot(data = health, x = '흡연상태', y = '혈색소', ci=None)
```

#####**인사이트 2 정리**
(흡연상태와 혈색소 간의 관계)

혈색소: 혈중의 산소운반체로서의 역할 수행

*  상자그림과 선 그래프를 통해 확인한 결과, **흡연자일수록 혈색소 수치가 높음**을 확인할 수 있다. 
*   흡연여부는 총 3가지 카테고리로 나눠진다. No(피우지 않는다), NotNow(이전에 피웠으나 끊었다), Yes(현재도 피우고 있다)

*   상자그림을 볼 경우, Yes 그룹이 훨씬 혈색소 범위가 높고 그 다음으로는 NotNow(이전에 피웠으나 끊었다), No(피우지 않는다) 순이다. 
*   선 그래프로 확인하였을 때, No에서 Yes 그룹으로 넘어갈 때, 그래프가 크게 상승하였고 Yes에서 NotNow로 갈 때, 소폭 감소하였다. 





###**인사이트 3: 수축기혈압과 이완기혈압 간의 관계**

```python
sns.scatterplot(data = health, x = '수축기혈압', y = '이완기혈압')
```

```python
sns.scatterplot(data = health, x = '수축기혈압', y = '이완기혈압', hue = '연령대코드(5세단위)', palette='Paired').set(xlim = (130, 200), ylim = (80, 120))
```

```python
sns.scatterplot(data = health, x = '수축기혈압', y = '이완기혈압', hue = '성별코드').set(xlim = (130, 200), ylim = (80, 120))
```

```python
sns.boxplot(data = health, x = '수축기혈압', y = '이완기혈압', palette='Paired')
```

```python
sns.set(rc={'figure.figsize':(6, 5)}) 

sns.lineplot(data = health, x = '수축기혈압', y = '이완기혈압', ci=None,  hue = '성별코드').set(xlim = (130, 200), ylim = (80, 120))
```

```python
sns.set(rc={'figure.figsize':(6, 5)}) 

sns.lineplot(data = health, x = '수축기혈압', y = '이완기혈압', ci=None, hue = '연령대코드(5세단위)', palette='Paired').set(xlim = (130, 200), ylim = (80, 120))
```

#####**인사이트 3 정리**
(수축기혈압과 이완기혈압 간의 관계)



*   선 그래프, 산점도, 상자그림으로 확인해보았을 때, 수축기 혈압과 이완기 혈압은 **양의 상관관계**를 보임을 확인할 수 있다. 
*   즉, 수축기 혈압이 높은 사람인 경우, 이완기 혈압 또한 높음을 확인할 수 있다. 


*   수축기 혈압이 130mmHg보다 높고 이완기 혈압이 80mmHg보다 높으면 심혈관 질환 발생률 역시 비례하여 증가한다. 
*   선그래프에서 연령대 코드를 기준으로 분석을 하였다. 수축기 혈압이 130mmHg보다 높고 이완기 혈압이 80mmHg보다 높은 그룹만 모아봤을 때, 9와 11그룹이 많음을 확인할 수 있다. 
* 즉, 40-44세, 50-54세에서 심혈관 질환 발생률이 높다. 
* 또한, 선 그래프에서 성별을 기준으로 확인하였을 때, 여성의 비율이 높음을 확인할 수 있다. 즉, 여성이 남성에 비해 심혈관 질환 발생률이 더 높다.
